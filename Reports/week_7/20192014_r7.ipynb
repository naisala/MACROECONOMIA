<span style='font-family:"Times New Roman"'> The objective of the article is to answer the question of what were the causes of chronic inflation in Peru during 1970 and 1980. The authors take a historical approach to explain the effects of monetary and fiscal policies applied throughout of five governments: the government of Belaunde, Velasco, Bermúdez, García and, finally, Fujimori. For this, they also rely on quantitative data on economic growth, the effects of monetary and economic policies.

<span style='font-family:"Times New Roman"'> Given the historical approach of the text, along with the situation characterized by the application of bad economic policies, the authors provide important data from the international scenario to make a more complete analysis of the causes of inflation in the country. They mention that negative effects on the Peruvian economy due to the increase in world interest rates during 1980. They also highlight the fact that important sectors for the Peruvian economy were very dependent on foreign countries due to the high degree of interdependence, among these sectors mining stood out, a sector that was affected because the prices of minerals were established by the world market.
One of the most common characteristics of economics papers is usually the absence of personal evaluations. Economists tend to work more with objective and quantitative data, which limits their ability to offer subjective information. However, in this research, while the authors described the national macroeconomic history, a personal assessment is evident when they point out the intention of the policies applied in Velasco’s government. This can definitely be a double-edged sword, but for research purposes it is interesting to assess what the government's initial objectives were (a priori) and contrast it with the final results. It shows that, although an economic policy may be well intentioned, this does not imply that the results will be beneficial, there is always the possibility that the opposite effect will happen.

<span style='font-family:"Times New Roman"'> Among the weaknesses, it seems that the authors omit to describe the short-term negative effects of the neoliberal policies applied during the Fujimori government. It would have been pertinent to point out that it is the lowest social sectors that are most affected during the first years, as well as those policies that cause the increase of inequalities in the long term. In addition, the authors give a lot of weight to politics, but it would have been wise to delve a little more into the social dimension because, after the harmful effects on the Peruvian economy by previous governments, the people changed their perspective on state intervention in the economy, which was reflected in Fujimori’s government.

<span style='font-family:"Times New Roman"'> The previous point even opens the doors to future research on the impact of the citizen's perspective on the economy on the country's macroeconomic health. For which it would be convenient to analyze political trends such as populism. Likewise, the paper encourages more research on international events, for which it would be necessary to carry out further research on the impact of these events and their impact on the Peruvian economy (always emphasizing the external part). Here it would be important to describe the role of certain International Organizations such as the IMF to induce internal measures to be adopted.

<span style='font-family:"Times New Roman"'> Finally, much has been said about the economic consequences of the governments of Belaunde, Velasco, Bermúdez, García and Fujimori in books that encompass social, political and economic dimensions, focusing more on the agent. But in this case, it is a specialized article on the macroeconomic part and its accumulated effects from one government to another until reaching the most critical period in modern Peruvian history: hyperinflation. The authors do a good job of analyzing all the periods from explaining why the policy makers made those decisions to their long-term consequences.

<span style='font-family:"Times New Roman"'>The slope of the IS curve is sensitive to changes in taxes, Marginal propensity to consume and Marginal propensity to import.  In this case, while the interest rate stays the same, when the tax rate increases the slope of the IS curve also increases, what finally causes a shift to the left in the IS_1 curve.

### <span style='font-family:"Times New Roman"'> Equilibrium in the money market

```python
plt.xlim(0,20)
plt.ylim(0,450)

plt.plot([3,17],[400,100],color='tab:green', label = "$M^d$")
plt.plot([10,10],[0,400],color='tab:grey', label = "$M^so$/Po")

plt.plot([10,0],[250,250],color='lightgrey',linestyle='--', label = "$r^eo$")

plt.title("Equilibrium in the money market")
plt.legend(frameon = False)
plt.xlabel("$M^s$/P")
plt.ylabel("$r$")
```

<span style='font-family:"Times New Roman"'>Following an increase in the level of income, the demand curve shifts to the right, causing the demand for money to become greater than the supply of money. In that case, the interest rate, which is the price of money, would have to be raised. Otherwise, when the level of income decreases, the demand curve ($M^d$) contracts to the left, which makes the demand less than the supply. This means that the interest rate must be lowered to "clean" that money.
Furthermore, if money is increased without reasoning, the interest rate (r) falls, which drives away foreign investment. If the interest rate falls, we also drive away investment in dollars. If dollars are scarce, the exchange rate shoots up.

### <span style='font-family:"Times New Roman"'> The Equilibrium LM Curve in the Money Market

```python
"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
c1 = 1000              # Precautionary money demand
c2 = 10                # Transaction money demand
c3 = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE LM-SCHEDULE"
def i_LM(c1, c2, c3, Ms, P, Y):
    i_LM = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_LM

i = i_LM(c1, c2, c3, Ms, P, Y)
    
"4|PLOT THE IS-SCHEDULE"
y_max = np.max(i)

v = [0, Y_size, 20, 90]                       # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="The Equilibrium LM Curve in the Money Market", xlabel=r'Y', ylabel=r'r')
plt.plot([0,0],[0,0], color='k', label = "LM")
ax.plot(Y, i, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks


plt.axis(v)                                     # Use 'v' as the axes range
plt.show()
```

<span style='font-family:"Times New Roman"'>In the Money Market, the relationship between the variables is direct (as one increases, the other increases). If we increase the level of income, the interest rate increases, which causes the demand for money to expand. With increases in the level of income in the Money Market, the interest rate also increases. Therefore, LM is made up of all the equilibrium points in the Money Market. When the actual money supply increases, the LM curve shifts to the right. In the opposite case, if the level of income does not increase, but more money is injected, the interest rate falls.
