**Using IS and LM equations you have to find the $Y^e$ and $r^e$. Please show all the steps you follow to arrive at the equilibrium equations**

The IS curve is given as:

<center>$r = \frac{\beta_0}{h} - \frac{\beta_1}{h}Y$<center>

Where:

<center>$ùõÉ_ùüé = ùêÇ_ùüé + ùêà_ùüé + ùêÜ_ùüé + ùêó_ùüé$  y  $ùõÉ_ùüè = ùüè ‚àí (ùêõ ‚àí ùê¶)(ùüè ‚àí ùê≠)$<center>

Also, the LM curve is given as: 

<center>$r = -\frac{1}{j}\frac{M^s_0}{P_0}+\frac{k}{j}Y$<center>

This two curves equations need to be equilized in order to find 
the equilibrium equations of $Y^e$ and $r^e$. In this exercise we see how the exogenous variables (r and Y) affect the endogenous variables.

1. Equilibrium level of income: we equilize interest rates

<center>$-\frac{1}{j}\frac{M^s_0}{P_0}+\frac{k}{j}Y=\frac{\beta_0}{h} - \frac{\beta_1}{h}Y$<center>

<center>$Y^e=\frac{j\beta_0}{kh+j\beta_1}+\frac{h}{kh+j\beta_1}\frac{M^s_0}{P_0}$<center>

2. Equilibrium interest rate: we equilize levels of income

<center>$r^e=\frac{k\beta_0}{kh+j\beta_1}-\frac{\beta_1}{kh+j\beta_1}\frac{M^s_0}{P_0}$<center>

Solving by any of the three methods: reduction, equalization or substitution, the values of the income and the interest rate that balance both markets are finally obtained.

**Simultaneous equilibrium in the goods and money markets**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
# IS Curve
Y_size = 100 
b = 0.4                 # Marginal propensity to consume
m = 0.2                 # Marginal propensity to import
t = 0.9                 # Taxes
Co = 20                 # Autonomous consumption
Io = 10                 # Investment intercept (when i = 0)
Go = 50                 # Government spending 
Xo = 2                  # Exports (given) 
h =  1                  # Investment slope wrt to i

Y = np.arange(Y_size)   # Array values of Y
#-------------------------------------------------------
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
#-------------------------------------------------------
# LM Curve
Y_size = 100

k = 15                # Transaction money demand
j = 10                # Speculation money demand
Ms = 9000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)
#-------------------------------------------------------
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
#-------------------------------------------------------
y_max = np.max(i)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, r, "k-", label = "IS", color = "yellowgreen")
ax.plot(Y, i, "b-", label="LM", color = "firebrick")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.52, linestyle='dashed', color = 'lightgrey')
plt.axhline(y=32, xmin= 0, xmax= 0.52, linestyle='dashed', color = 'lightgrey')
plt.plot(51, 32, marker="o", color="black", label = "E")
plt.text(50,38, 'E', fontsize = 14, color = 'black')

ax.legend()
plt.show()
```

## DA ‚àí OA

**Derive the DA equation from:**

The Aggregate Demand and Supply Model, DA - OA, incorporates the IS-LM model equations (with which Aggregate Demand is determined) and the Aggregate Supply equation.

* **Aggregate Demand (AD)**

   - IS: $[1-b(1-t)]Y = C_0 + I_0 + G_0 - hr$

- LM: $\large \frac{M^s_0}{P_0}$ $= kY - jr$

-  LM(B): $M^s_0 - P = kY - jr$

- $P$ = $\large \frac{hM^s_0 + j \beta_0}{h} - \frac{j \beta_1 + hk}{h}Y$

* **Aggregate Supply (AS)**

<center> $P = P^e + \Theta(Y ‚àí \bar{Y})$ where $\bar{Y} = Y_f$</center>

$ \bar{Y} \dashrightarrow$ Potential Output

$ P^e \dashrightarrow$ Expected Price Level

$\mathbf{Intuition:}$

An increase in the potential product causes the opposite effect on the price level, and this directly affects the money demand, which will increase so that interest rate needs to decrease. This is positive for the investment because will get increased. On the other hand, the aggregate demand will expand, even more than the production level, at it will make the product to increase.

<center> $P\downarrow \rightarrow M^o\uparrow \rightarrow M^o>M^s_0 \rightarrow r\downarrow \rightarrow I_0\uparrow \rightarrow AD\uparrow \rightarrow AD>Y \rightarrow Y\uparrow$<center> 

<center>$\bar{Y}\uparrow \rightarrow P\downarrow$<center> 

$\mathbf{Math:}$

- $\frac{dY}{d\bar{Y}} = \frac{-\Theta h M^s_0}{|A|P^2_0} > 0$


- $\frac{dr}{d\bar{Y}} = \frac{\Theta j(1-b(1-t))+\Theta hk}{|A|} < 0$


- $\frac{dP}{d\bar{Y}} = \frac{\Theta(1-b(1-t))M^s_0}{|A|P^2_0} < 0$
