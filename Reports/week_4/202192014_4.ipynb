**1) Reading Part**

Colonialism and the world economy 

The article seeks to answer a simple but essential question for the world economy, and that is, how can we explain the different variations of GDP per capita in all the countries around the world? To do this, the authors go back to the time of colonialism and argue that it is mortality rates that influence the decision colonizing countries make to choose whether to settle there or not, this in turn leads to an impact on the type of institutions they create and finally in economic performance.

The research itself is a double-edged sword, on one hand, one of its strengths is that to explain the economic performance of countries they currently go back many years and focus, from a historical perspective, on explaining the variable of mortality rate to explain it, checking that this explanatory variable and others do not have a spurious relationship, that is, they take care that mortality does not directly affect the institutions but only the settlement. On the other hand, going back so many years and focusing on the mortality rate variable can be counterproductive because during that period of time there is a potential probability that events of another political, economic, military, etc. nature have occurred that have also affected the decision to creation of extractive institutions or the performance of current economies. In addition, because a large number of countries are taken into account to make the generalization, it is very likely that these other local events that have not been taken into account were even more significant than the main variable of the study.

Although the text may fall short to take into account other variables and the local events of each country, it is important to highlight the efforts of the authors to generate a general hypothesis that can be applied in the world to explain not only the better performance of some countries like the United States or Canada but also the lower performance of countries close to the equator. This kind of structuralist explanation helps us to identify what are those historical problems that we have been dragging from the past to the present and creates the way to help future research in our region to create hypothetical solutions to these problems.

The text has taken the first step and has focused on the explanation of a variable in just 30 pages, the next step could be to expand the investigation and add other types of variables to the model that are significant when explaining economic performance. of the countries. It could even be separated into different causes for each region, maintaining the historical perspective.

**2) Code section**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/naisala/MACROECONOMIA/main/peru_gdp_1922_2021.csv"
```

```python
base = pd.read_csv(url)
```

```python
print(base)
```

```python
base = base.rename( columns = { "Unnamed: 0":'Año', "PM04862AA": 'PBI'})
base
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x = "Año",y = "PBI", data = base)
plt.title("PBI per capita Perú 1922-2021")
```

```python
base['Creci_Anual'] = base["PBI"].pct_change()*100
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x = 'Año', y= 'Creci_Anual', data=base, color = "orange")
plt.axhline(y = base.Creci_Anual.mean(), linestyle = '-', label = "Crecimiento Promedio", color = "red")
plt.xlabel("Años")
plt.ylabel("Tasa de crecimiento del PBI")
plt.title("Tasa de crecimiento del PBI del Perú 1922-2021")
```

The average annual growth rate of Peru's GDP between 1922 and 2021 has remained positively between 0% and 5%. In the graph, two periods stand out: from 1985 to 1990, in which GDP growth fell during the term of former President Alan García, characterized by hyperinflation; From the 90s to the 2000s, an increase in GDP growth also began to be experienced during the mandate of Alberto Fujimori, a growth caused by the liberal policies (economic liberalization) of the country.
