The article formulates a macroeconomic model as a result of the COVID-19 pandemic and asks if it is capable of reproducing the impact of the quarantine in a two-sector economy: sector 1, the one directly affected by the quarantine and which encompasses the hotels, restaurants, airlines, etc.; while sector 2, the one indirectly affected, is characterized by the production of food and essential goods and services. Furthermore, both sectors are interrelated because workers consume goods from both.

One of the strengths of this research is the proposed model. Its structure based on the division of an economy into two sectors serves as a template not only for health crises but also for other types of crises that may occur in the future, one of the most common in Peru being natural disasters such as the "Fen√≥meno del Ni√±o". Within the weaknesses, it would have been important if the authors had introduced the issue of the elasticity of certain products to explain the sensitivity of sector 1 compared to sector 2. In addition, during the quarantine and due to COVID-19 there were elastic goods that became be inelastic and, therefore, affected the composition of goods in both sectors. Like alcohol or masks, which have become essential goods. Likewise, the authors assume for their equations that the actors are based on the previous prices to make a decision because their incomes fell. However, that thesis was not adjusted to what was happening during the pandemic. Far from reducing their consumption, the actors rushed to buy basic necessities en masse, including toilet paper, a collective euphoria triggered by the crisis. This is where the concept of rational expectations proposed by Robert Lucas would be introduced, that when actors make economic decisions in the present they are doing so thinking about the future and not the past.

The study of the economic problems triggered by the quarantine shows, by comparing two exercises, what were the effects produced on the level of production and the level of prices. **Derive mathematically the equilibrium in the IS-LM model. You have to find the equilibrium equations of $Y^e$ and $r^e$.**

The IS curve is given as:

<center>$r = \frac{\beta_0}{h} - \frac{\beta_1}{h}Y$<center>

Where:

<center>$ùõÉ_ùüé = ùêÇ_ùüé + ùêà_ùüé + ùêÜ_ùüé + ùêó_ùüé$  y  $ùõÉ_ùüè = ùüè ‚àí (ùêõ ‚àí ùê¶)(ùüè ‚àí ùê≠)$<center>

Also, the LM curve is given as: 

<center>$r = -\frac{1}{j}\frac{M^s_0}{P_0}+\frac{k}{j}Y$<center>

This two curves equations need to be equilized in order to find 
the equilibrium equations of $Y^e$ and $r^e$. In this exercise we see how the exogenous variables (r and Y) affect the endogenous variables.

1. Equilibrium level of income: we equilize interest rates

<center>$-\frac{1}{j}\frac{M^s_0}{P_0}+\frac{k}{j}Y=\frac{\beta_0}{h} - \frac{\beta_1}{h}Y$<center>

<center>$Y^e=\frac{j\beta_0}{kh+j\beta_1}+\frac{h}{kh+j\beta_1}\frac{M^s_0}{P_0}$<center>

2. Equilibrium interest rate: we equilize levels of income

<center>$r^e=\frac{k\beta_0}{kh+j\beta_1}-\frac{\beta_1}{kh+j\beta_1}\frac{M^s_0}{P_0}$<center>

Solving by any of the three methods: reduction, equalization or substitution, the values of the income and the interest rate that balance both markets are finally obtained.

**Replicate, explain with math and intuition the next figure:** Simultaneous equilibrium in the goods and money markets

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
# IS Curve
Y_size = 100 
b = 0.4                 # Marginal propensity to consume
m = 0.2                 # Marginal propensity to import
t = 2.6                 # Taxes
C0 = 20                 # Autonomous consumption
I0 = 10                 # Investment intercept (when i = 0)
G0 = 50                 # Government spending
X0 = 2                  # Exports (given)
h =  1                  # Investment slope wrt to i
Y = np.arange(Y_size)   # Array values of Y

def r_IS(b, m, t, C0, I0, G, X0, h, Y):
    r_IS = (C0 + I0 + G0 + X0 - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, C0, I0, G0, X0, h, r):
    Y_IS = (C0 + I0 + G0 + X0 - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, C0, I0, G0, X0, h, Y)

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

y_max = np.max(i)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Simultaneous equilibrium in the goods and money markets", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, r, "k-", label = "IS", color = "red")
ax.plot(Y, i, "b-", label="LM", color = "blue")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.52, linestyle='dashed', color = 'lightgrey')
plt.axhline(y=15, xmin= 0, xmax= 0.52, linestyle='dashed', color = 'lightgrey')
plt.plot(51, 15, marker="o", color="black", label = "E")
plt.text(50,18, 'E', fontsize = 14, color = 'black')

ax.legend()
plt.show()
```

The point E ($E=Y_0, r_0$) represents the point of intersection where both markets are in equilibrium, there is no excess demand or supply in both the money market and the goods market.

**Fiscal Policy**

IS Curve: If autonomous spending rises ($G_0$), aggregate demand increases ($DA$), this is greater than the level of production and therefore, Y rises.
<center>$G_0\uparrow \rightarrow DA\uparrow \rightarrow DA>Y \rightarrow Y\uparrow$
    
LM Curve: If the level of production rises ($Y$), the demand for money rises ($M^d$), this makes it greater than the supply ($M^d > \frac{M^s}{P}$) and therefore the interest rate (price of money = $r$) increases.
<center>$Y\uparrow \rightarrow M^d\uparrow \rightarrow M^d > \frac{M^s}{P} \rightarrow r\uparrow$

**Expressed mathematically:**

$\Delta Y= [\frac{j}{kh+j\beta_1}] \Delta G_0>0$

$\Delta r= [\frac{k}{kh+j\beta_1}] \Delta G_0>0$

**Expansionary Monetary Policy**

With this policy, the supply of money increases, which makes it greater than the demand for money, which also causes the interest rate to fall. 

<center>$M^o\uparrow \rightarrow M^o>M^d \rightarrow r\downarrow$
    
When interest rates gets reduced, it causes investment to raise as well as aggregate demand, which also makes this last to be greater than production level.

<center>$r\downarrow \rightarrow I_o\uparrow \rightarrow DA\uparrow \rightarrow DA>Y \rightarrow Y\uparrow$

**Restrictive Fiscal Policy**

If autonomous spending falls, aggregate demand falls with it, which makes the latter less than the level of production, so it also falls. 

<center>$G_0\downarrow \rightarrow DA\downarrow \rightarrow DA<Y \rightarrow Y\downarrow$<center>   

If the level of production also fell, the demand for money would also fall, that makes the demand for money less than the supply of money and finally, the interest rate must fall
    
<center>$Y\downarrow \rightarrow M^d\downarrow \rightarrow M^d < M^o \rightarrow r\downarrow$<center>

**Expressed Mathematically**

Expansionary Monetary Policy

$\Delta Y= [\frac{h}{kh+j\beta_1}]\frac{1}{P_0} \Delta M^s_o>0$

Restrictive Fiscal Policy

$\Delta Y= [\frac{j}{kh+j\beta_1}] \Delta G_0>0$

$\Delta r= [\frac{k}{kh+j\beta_1}] \Delta G_0>0$
